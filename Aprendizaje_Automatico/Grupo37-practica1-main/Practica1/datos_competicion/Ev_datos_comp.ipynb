### Estudio sobre la media y la desviación típica

```python
# Analisis estadístico
# ==============================================================================
sub_df = comp_df.iloc[:, :76]

# Calcular el número de grupos
num_grupos = sub_df.shape[1] // 5


# Dividir el subconjunto en grupos de 5 variables cada uno
grupos = [sub_df.iloc[:, i:i+5] for i in range(1, sub_df.shape[1], 5)]

# Calcular las medias y desviaciones típicas de cada grupo
medias = [grupo.mean() for grupo in grupos]
desviaciones = [grupo.std() for grupo in grupos]

# Crear un gráfico diferente para cada grupo
for i in range(len(grupos)):
    fig, axs = plt.subplots(2, 1, figsize=(8, 8))
    
    # Graficar la media en el primer subgráfico
    axs[0].errorbar(grupos[i].columns, grupos[i].mean(), yerr=grupos[i].std(), fmt='o', capsize=5)
    axs[0].plot(grupos[i].columns, grupos[i].mean(), 'o--')
    axs[0].set_title(f'Grupo {i+1}' , fontweight='bold' )
  
    axs[0].set_ylabel('Media', weight='bold')
    axs[0].tick_params(axis='x', rotation=90)
    # Graficar la desviación típica en el segundo subgráfico
    axs[1].errorbar(grupos[i].columns, grupos[i].std(), fmt='o', capsize=5)
    axs[1].plot(grupos[i].columns, grupos[i].std(), 'o--')
  
    axs[1].set_ylabel('Desviación típica' ,weight='bold')
    axs[1].tick_params(axis='x', rotation=90)
    plt.tight_layout()
    plt.show()
    # Crear un diccionario con los índices y las diferencias de medias de cada grupo
diff_medias = {}
for i, grupo in enumerate(grupos):
    diff_medias[i+1] = abs(grupo.mean().max() - grupo.mean().min())

# Ordenar los grupos según las diferencias de medias en orden descendente
sorted_diff_medias = sorted(diff_medias.items(), key=lambda x: x[1], reverse=True)

# Imprimir los 3 grupos con las mayores diferencias de medias
print("Los 3 grupos con las mayores diferencias de medias son:")
for i in range(3):
    grupo = sorted_diff_medias[i]
    print(f"Grupo {grupo[0]} con una diferencia de medias de {grupo[1]:.2f}")

# Crear un diccionario con los índices y las diferencias de desviaciones típicas de cada grupo
diff_desviaciones = {}
for i, grupo in enumerate(grupos):
    diff_desviaciones[i+1] = abs(grupo.std().max() - grupo.std().min())

# Ordenar los grupos según las diferencias de desviaciones típicas en orden descendente
sorted_diff_desviaciones = sorted(diff_desviaciones.items(), key=lambda x: x[1], reverse=True)

# Imprimir los 3 grupos con las mayores diferencias de desviaciones típicas
print("Los 3 grupos con las mayores diferencias de desviaciones típicas son:")
for i in range(3):
    grupo = sorted_diff_desviaciones[i]
    print(f"Grupo {grupo[0]} con una diferencia de desviaciones típicas de {grupo[1]:.2f}")
    # Crear una lista vacía para almacenar los grupos con desviación típica cero
grupos_con_desviacion_cero = []

# Iterar sobre cada grupo
for i in range(len(grupos)):
    # Verificar si alguna de las desviaciones típicas es igual a cero
    if any(grupos[i].std() == 0):
        # Si es así, agregar el índice del grupo a la lista de grupos con desviación cero
        grupos_con_desviacion_cero.append(i+1)

# Imprimir los grupos con desviación típica cero
print("Grupos con desviación típica cero:", grupos_con_desviacion_cero)
for i in range(len(grupos)):
    grupo_con_desv_cero = grupos[i].loc[:, grupos[i].std() == 0]
    if not grupo_con_desv_cero.empty:
        print(f"En el grupo {i+1}, la desviación típica es cero para el atributo {grupo_con_desv_cero.columns[0]}.")
```

# MODELO FINAL
----------------------
Una vez establecido el mejor modelo, entrenarlo y guardarlo en un archivo '.pkl', nos disponemos a emplearlo para los datos de competición. Para ello, importamos el archivo, predecimos la salida y guardamos dichas predicciones en un archivo denominado 'predicciones.csv'.

```python
import pickle
import os
import csv

# Abrir el archivo del modelo
with open('../datos_disponibles/modelo_final.pkl', 'rb') as f:
    modelo_final = pickle.load(f)

# Predecimos con el modelo final - Ridge
# ==============================================================================
predicciones_numpy = modelo_final.predict(x_comp_numpy)

# Guardamos las predicciones en un archivo csv
# ==============================================================================
csv_filename = "predicciones.csv"
with open(csv_filename, 'w') as file:
    escritor_csv = csv.writer(file, delimiter=",")
    
    # Escribimos los datos
    for fila in predicciones_numpy:
        escritor_csv.writerow(fila)

# Cerramos el archivo
file.close()
```
